```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
from deepimpute.multinet import MultiNet
import os
from tqdm.notebook import tqdm

import numpy as np
import sys
sys.path.append("/home/humble_local2_25t/alexw/MIST_round3/MIST_additional_holdout/scripts/")
from evalHO import evalSlide
from utils import data_norm
```

```python
prj_fd = "/home/humble_local2_25t/alexw/MIST_round3/MIST_additional_holdout/data"
data_names = os.listdir(prj_fd)
data_names.remove( 'spot_metadata.tsv')
data_names
```

```python
for i, dn in enumerate(tqdm(data_names)):
    for j in range(5):
        ho_fn = f"{prj_fd}/{dn}/ho_data_{j}.csv"
        df = pd.read_csv(ho_fn, index_col=0)
        df, _ = data_norm(df, method='cpm')
        model = MultiNet()
        model.fit(df)
        imputed = model.predict(df)
        imputed = imputed.loc[df.index, df.columns]
        out_fd = f"{prj_fd}/{dn}/imputed/"
        
        if not os.path.exists(out_fd):
            os.mkdir(out_fd)
            
        imputed.to_csv(f"{out_fd}DeepImpute_{j}.csv")
```
